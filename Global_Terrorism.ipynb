## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

#I required to do two types of grouping. In first I required to count the desired event and in second one, the sum of values was required.
#Hence, I made two group by functions.

def Grouping(df,second,*first): # *is used so that I could use multiple inputs.

  '''This function counts the occuerence of desired event.
     first and second is the actual column names'''

  A=df.groupby([*first])[second].count() #Count gives Idea about the frequency or repition.
                                         # It clarifies the idea about number of attacks or number of events occured.
  B=pd.DataFrame(A)                      # Creation of dataframe will help for easy plotting
  return B

def Grouping_sum(df,second,*first):      #This function helps for summing all the values. It helps to count total number of kills.
  '''df is the desired dataframe          
     first and second is the name of desired columns'''   

  A=df.groupby([*first])[second].sum()
  B=pd.DataFrame(A)
  return B

def Rename(var,first,second,third,fourth,fifth,sixth):  #After using above function,proper renaming comes through this function.
  '''var=the name of variable
     first and second is the actual column name 
     and the new column name respectively.'''
  #This function is useful for, three columns dataframe.

  new_var=var.rename(columns={first:second}).reset_index()
  A=new_var.rename(columns={third:fourth,fifth:sixth})
  return A
```

```python
def Maximum_values(df,column1 , column2):
  
  dataframe_list= []
  for i in df[column1].unique():

    first= df[df[column1]== i ]

    second = first[first[column2] == first[column2].max()]
    third = pd.DataFrame(second)
    dataframe_list.append(third)

  first_object = dataframe_list[0]
  second_object = pd.concat([first_object,dataframe_list[1]],axis = 0)
  for i in dataframe_list[2:]:
    second_object

    second_object = pd.concat([second_object,i],axis = 0)
  

  third_target=second_object.rename(columns={column2:'Max'+' '+ column2})
  return third_target
```

```python

limits = [(0,10),(10,31),(31,39),(39,46),(46,72),(72,95),(95,98),(98,112),(112,121),(121,133),(133,182),(182,203)]
colors1 = ['rgb(167, 119, 12)',
 'rgb(197, 96, 51)',
 'rgb(217, 67, 96)',
 'rgb(221, 38, 163)',
 'rgb(196, 59, 224)',
 'rgb(153, 97, 244)',
 'rgb(95, 127, 228)',
 'rgb(40, 144, 183)',
 'rgb(15, 151, 136)',
 'rgb(39, 153, 79)',
 'rgb(119, 141, 17)',
 'rgb(167, 119, 12)',
 '#2d47f9',
 '#2593ff'] 
cities = []
scale = 5000

fig = go.Figure()

for i in range(len(limits)):
    lim = limits[i]
    df_sub = choro_df[lim[0]:lim[1]]
    fig.add_trace(go.Scattergeo(
        locationmode = "country names" ,
        lon = df_sub['longitude'],
        lat = df_sub['latitude'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['Country_count']/10,
            color = colors1[i],
            line_color='rgb(40,40,40)',
            line_width=0.5,
            sizemode = 'area'
        ),
        name = df_sub['Region'][lim[0]]))

fig.update_layout(
        title_text = 'Countries and Number of Terror Attacks.',
        showlegend = True,
        geo = dict(
            scope = 'world',
            landcolor = 'rgb(217, 217, 217)'
        )
    )
fig.update_layout(width=1200,height=900)
fig.show('jpg')
```
